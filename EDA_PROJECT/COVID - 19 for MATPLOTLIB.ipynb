# Making Line Graph using PLOT Function

```python
plt.figure(figsize=(23,15))
plt.plot(x,y,marker = "o",color = "hotpink", mec = "green")
plt.title("Covid Deaths and Active Cases")

plt.show()
```

```python
x1  = cvd["Country/Region"].head(10)
y1 = cvd["Recovered"].head(10)
plt.figure(figsize=(23,15))
plt.plot(x1,y1,marker = "o",color = "#FF0000", mec = "green")
plt.title("Covid Deaths and Active Cases")
plt.xlabel("COVID-19 Death Cases")
plt.ylabel("COVID-19 Active Cases")
plt.show()
```

```python
plt.figure(figsize=(28,15))
plt.plot(x1,y1,marker = "o")
plt.show()
```

# Changing Marker Edge Color using mec parameter

```python
plt.figure(figsize=(20,15))
plt.plot(x1,y1,marker = "o",color = "red", mec = "green")
plt.show()
```

```python
p = cvd["WHO Region"].head(10)
q = cvd["Active"].head(10)
plt.figure(figsize=(20,15))
plt.plot(p,q,marker = "o",mec = "green")
plt.show()
```

```python
plt.plot(p,q,marker="o",ms="12", color="hotpink",linewidth = "3.5",mec="r",mfc = "g")
plt.show()
```

# Using Grid Function

```python
plt.plot(p,q,marker = "o", mec = "green")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,15))
plt.plot(x1,y1,marker = "o", mec = "green")
plt.grid()
plt.show()
```

# Using Grid Function and Implementing axis as X to give it Verticle grid

```python
plt.figure(figsize=(20,15))
plt.plot(x1,y1,marker = "o",mec = "red")
plt.grid(axis = "x")
plt.show()
```

# Using Grid Function and Implementing axis as Y to give it Horizontal grid

```python
plt.figure(figsize=(20,15))
plt.plot(p,q,marker = "o",mec = "red")
plt.grid(axis = "y")
plt.show()
```

# GRID linestyle as Dashed
